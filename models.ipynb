## Three Models To Classify

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
df = pd.read_csv("heart.csv")
df.sample()
```

```python
# df = pd.get_dummies(df, columns = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal'])
```

```python
# ss = StandardScaler()
# columns_to_scale = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
# df[columns_to_scale] = ss.fit_transform(df[columns_to_scale])
```

```python
x = df.iloc[:,0:-1]
y = df.iloc[:,-1]

# x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']),  # Scale numerical columns
        ('cat', OneHotEncoder(drop='first'), ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal'])    # One-hot encode categorical columns
    ]
)
```

## KNearestNeighbor

```python
knn = KNeighborsClassifier()

# Create a pipeline
pipeline1 = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', knn)
])

# Fit the pipeline
pipeline1.fit(x, y)
```

## Support Vector Classifier

```python
svc_classifier = SVC(kernel = 'linear')

# Create a pipeline
pipeline2 = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', svc_classifier)
])

# Fit the pipeline
pipeline2.fit(x, y)
```

## Decision Tree Classifier

```python
dt_classifier = DecisionTreeClassifier(max_features = 3, random_state = 0)

# Create a pipeline
pipeline3 = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', dt_classifier)
])

# Fit the pipeline
pipeline3.fit(x, y)
```

### Exporting the models using pickle

```python
import pickle
with open('knn_pipeline.pkl', 'wb') as f:
    pickle.dump(pipeline1, f)

print("Pipeline1 saved successfully!")
```

```python
with open('svc_pipeline.pkl', 'wb') as f:
    pickle.dump(pipeline2, f)

print("Pipeline2 saved successfully!")
```

```python
with open('dt_pipeline.pkl', 'wb') as f:
    pickle.dump(pipeline3, f)

print("Pipeline3 saved successfully!")
```
